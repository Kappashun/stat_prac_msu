```python
import pandas as pd
import numpy as np
from collections import defaultdict
from sklearn.cluster import KMeans, DBSCAN, SpectralClustering, MeanShift

df = pd.read_csv('~/datasets/Table.csv'
                ).drop(columns=["Github repo", "Group", "Student.1"]
                ).fillna(0
                ).replace(',', '.', regex=True
                ).set_index("Student"
                ).iloc[1:]

columns = ['Pandas', 'kNN', 'Unnamed: 10', 'Unnamed: 13', 'Unnamed: 24', 'Unnamed: 28']
df = df[columns].apply(pd.to_numeric, errors='coerce')
features = df.values

kmeans = KMeans(n_clusters=4, random_state=42).fit(features)
kmeans = kmeans.fit(df.values)
df['k-means'] = kmeans.labels_
df['k-means'] = df['k-means'].map({0: 'нз', 1: 'уд', 2: 'хор', 3: 'отл'})

dbscan = DBSCAN(eps=0.5, min_samples=5).fit(features)
df['dbscan'] = dbscan.labels_
df['dbscan'] = df['dbscan'].map({0: 'нз', 1: 'уд', 2: 'хор', 3: 'отл'}).fillna('noise')

sc = SpectralClustering(4).fit(features)
df['spectral'] = sc.labels_
df['spectral'] = df['spectral'].map({0: 'нз', 1: 'уд', 2: 'хор', 3: 'отл'})

mean_shift = MeanShift().fit(features)
df['mean_shift'] = mean_shift.labels_
df['mean_shift'] = df['mean_shift'].map({0: 'нз', 1: 'уд', 2: 'хор', 3: 'отл'})
```
