### Problem 6
Justify (orally) that the accept-reject method really samples from the required distribution. Simulate $1000$ points from the distribution with density $e^xcos^2x$ on the segment $\left[-\dfrac\pi2, \dfrac\pi2\right]$. Plot the sampled histogram on the sample and compare it with the plot of the exact density function.

```python
fig = plt.figure()

x = np.linspace(-np.pi/2, np.pi/2, 100)
density = np.exp(x)*np.cos(x)**2 # not normalized density!
plt.fill_between(x, 0, density)
plt.ylabel('PDF (not normalized)')
plt.xlabel('x')
plt.show()
```

```python
# Your code here
def target_density(x):
    return np.exp(x) * np.cos(x)**2

def accept_reject(num_samples, max_pdf):
    sampled_points = []
    while len(sampled_points) < num_samples:
        x = np.random.uniform(-np.pi/2, np.pi/2)
        u = np.random.uniform(0, max_pdf)

        if u <= target_density(x):
            sampled_points.append(x)
    return np.array(sampled_points)


sampled_points = accept_reject(1000, 1.5)
norm_coef = 1.84

plt.figure(figsize=(8, 6))
plt.hist(sampled_points, bins=30, density=True, color='skyblue', edgecolor='black', alpha=0.7, label='Sampled Histogram')

x = np.linspace(-np.pi/2, np.pi/2, num_samples)
density = target_density(x) / norm_coef
plt.plot(x, density, color='red', linewidth=2, label='Exact Density Function')

plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Sampled Histogram from Distribution with Density e^x * cos(2x) using Rejection Sampling')
plt.legend()
plt.show()
```
